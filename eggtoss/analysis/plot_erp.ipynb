```python
experiment_name = 'e'
data_dir = f'./data/{experiment_name}'
exp = eeg_decoder.Experiment(experiment_name,data_dir,dev=False,trim_timepoints=[-200,1500])
# exp.load_artifact_idx(0);
# exp.load_behavior(3);
# exp.xdata_files = exp.xdata_files[:-3]
# exp.ydata_files = exp.ydata_files[:-3]
# exp.artifact_idx_files = exp.artifact_idx_files[:-3]
# exp.behavior_files = exp.behavior_files[:-3]
# exp.nsub = len(exp.xdata_files)
erp = eeg_decoder.ERP(exp,subtitle=experiment_name)
xdata_all, ydata_all = erp.load_all_eeg()
```

```python

fig,axes = plt.subplots(1,3,figsize=(19,4))
for isubset,(subtitle,subset) in enumerate(zip(['Parietal-Occipital','Central','Frontal'],[('P','O'),('C'),('F')])):
    erp.plot_ss(xdata_all, ydata_all,ax = axes[isubset], electrode_subset=subset, condition_subset=[11,12,21,22],
                condition_labels = ['1 color','2 color','1 motion','2 motion'])
    axes[isubset].set_title(subtitle)
plt.savefig('erp.png',dpi=500,bbox_inches='tight')
plt.show()
```
